# Data Visualization

```python
# Histograms it will used numerical columns only.
# histogram will show the outliers of the particular data in the dataset.
# It can compare the different part of data and detect the pattern and anomalies.

df.hist()
plt.tight_layout()# it will adjust all the numerical columns in one cell.
plt.show()
```

```python
# Boxplot will used to find the outliers in the columns in dataset.
# it can show the spread of data and skewness of the data.
# it can compare the multiple groups in single cell and also detect the outliers.
df.boxplot()
plt.tight_layout()
plt.show()
```

```python
# as we find that all the columns are have the outliers, but in most outliers are in Total sales value and net sales values.
```

```python
categorical_col = df.select_dtypes(include = "object")# it will show the all categorical columns in the dataset.
categorical_col
```

```python
# Barchart it will used only for categorical columns only.
# categories will be on X-axis and frequency will be on Y-axis.
# every unique item will show separtely.
# it will also find the outliers.

plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_col, 1):
    plt.subplot(3, 3, i)  # Adjust rows/cols depending on no. of columns
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"{col}")

plt.tight_layout()
plt.show()
```

```python
for col in categorical_col:
    print(df[col].value_counts())     # frequency of each category
    print(df[col].unique())
```

Standardization of Numerical Variables

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
print(df[numeric_cols].describe())
```

```python
# we can used the data Tranformer
# the standardization is the method which put the features on the same scaled.
sc = StandardScaler()
df_standardscaler = df.copy()
```

```python
df_standardscaler[numeric_cols] = sc.fit_transform(df[numeric_cols])
```

```python
print(df_standardscaler[numeric_cols].describe())
```
